```python
import os

def count_files(path):
    file_count = sum([len(files) for r, d, files in os.walk(path)])
    return file_count

def Dataset2():
    main_folder_path = "./images/"
    folders = os.listdir(main_folder_path)
    folder_file_counts = [(folder, count_files(os.path.join(main_folder_path, folder))) for folder in folders]
    sorted_folder_file_counts = sorted(folder_file_counts, key=lambda x: x[1], reverse=True)
    
    for folder, file_count in sorted_folder_file_counts:
        print(f"{folder}: {file_count}")
        
if __name__ == '__main__':
    Dataset2()
```

```python
import os

# Function to count the number of files in a directory
def count_files(path):
    count = 0
    for root, dirs, files in os.walk(path):
        count += len(files)
    return count

# Main directory path
main_dir_path = "./images/"

# Get the count of files in each subdirectory
sub_dir_counts = {}
for root, dirs, files in os.walk(main_dir_path):
    for dir in dirs:
        dir_path = os.path.join(root, dir)
        sub_dir_counts[dir_path] = count_files(dir_path)

# Sort the subdirectories by count of files in descending order
sorted_dirs = sorted(sub_dir_counts.items(), key=lambda x: x[1], reverse=True)

# Write the information of the top 30 directories in a text file
with open("top_20_foldersDataset2.txt", "w") as f:
    for i, (dir_path, count) in enumerate(sorted_dirs[:30]):
        f.write(f"{i+1}. {dir_path} - {count} files\n")
```
